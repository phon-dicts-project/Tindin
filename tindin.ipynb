```python
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
with open('tin_slovar.html', 'r', encoding='Windows-1251') as f: 
    text = f.read()
    
soup = bs(text,'html.parser')
post = soup.find_all('b', {'class': ''})
```

```python
word_list = []
word_dict = {}
#тут поменять на норм акуты
rome_nums = 'I'
for p in post:
    word = p.get_text()
    #print(word)
    if word.isupper() and word != rome_nums:
        word_list.append(word)
        word_dict['word'] = word_list
        print(word)
```

```python
df = pd.DataFrame(word_dict)
print(df)
```
