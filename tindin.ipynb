```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
```

```python
with open('tin_slovar.html', 'r', encoding='Windows-1251') as f: 
    text = f.read() 
    text = text.replace('µ','Ý')
    pattern1 = '<sup><.*>н</span></sup>'
    text = re.sub(pattern1, 'ᴴ', text)
    pattern2 = '''<p class=MsoNormal align=center style='text-align:center;text-indent:14.2pt'><b
style='mso-bidi-font-weight:normal'><span style='font-size:8.0pt;mso-bidi-font-size:
10.0pt'>.*<o:p></o:p></span></b></p>'''
    text = re.sub(pattern2, '', text)
    pattern3 = '''<p class=MsoNormal style='text-align:justify;text-indent:14.2pt'><i
style='mso-bidi-font-style:normal'><span style='font-size:8.0pt;mso-bidi-font-size:
10.0pt'><o:p>&nbsp;</o:p></span></i></p>'''
    text = re.sub(pattern3, '', text)
    pattern4 = '''<p class=MsoNormal style='text-align:justify;text-indent:14.2pt'><span
style='font-size:8.0pt;mso-bidi-font-size:10.0pt'><o:p>&nbsp;</o:p></span></p>'''
    text = re.sub(pattern4, '', text)
    pattern5 = '''<p class=MsoNormal style='text-align:justify;text-indent:14.2pt'><b
style='mso-bidi-font-weight:normal'><span style='font-size:8.0pt;mso-bidi-font-size:
10.0pt'><o:p>&nbsp;</o:p></span></b></p>'''
    text = re.sub(pattern5, '', text)
    pattern6 = '<sup>1</sup>'
    text = text.replace(pattern6, '¹')
    pattern7 = '<sup>2</sup>'
    text = text.replace(pattern7, '²')

    
```

```python
tin_dict = {}
lemm_list = []
morphon_list = []
trans_list = []
ws_list = []

soup = bs(text,'html.parser')
wordstrings = soup.find_all('p', {'style': 'text-align:justify;text-indent:14.2pt'})
for ws in wordstrings:
    strline = ''
    ws_list.append(ws)
    line = ws.get_text()
    wstr = line.split('\n')
    for i in wstr:
        strline += ' '
        strline += i
    line = strline
    lemm = line.split()[0]
    lemm_list.append(lemm)
    line = re.sub(lemm, '', line)
    trans_list.append(line)
    morphon = re.findall(r'\(.*?\)', line)
    morphon_list.append(morphon)
tin_dict['lemm'] = lemm_list
tin_dict['morphon'] = morphon_list
tin_dict['trans'] = trans_list
tin_dict['tag'] = ws_list
    
```

```python
df = pd.DataFrame(data=tin_dict)
df
```

```python
df.to_csv('tindin.csv', columns=['lemm', 'morphon', 'trans', 'tag'])
```
